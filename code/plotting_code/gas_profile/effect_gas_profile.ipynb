```python
from halo_model.power_spectra.param_effect import param_effect_spectrum
from halo_model.config.config import Config

sys.path.append(os.path.abspath("/home/milan/Desktop/thesis/code/plotting_code/gas_profile")) 

cfg = Config()
```

```python
fig1 = param_effect_spectrum(cfg, 'theta_ej', cfg.theta_ej - 2)
fig2 = param_effect_spectrum(cfg, 'delta', cfg.delta - 2)
fig3 = param_effect_spectrum(cfg, 'gamma', cfg.gamma - 0.5)
fig4 = param_effect_spectrum(cfg, 'logM_gaslimit', cfg.logM_gaslimit + 1)
fig5 = param_effect_spectrum(cfg, 'beta', cfg.beta - 0.5)
```

```python

k_grid = np.geomspace(cfg.k_min, cfg.k_max, cfg.N_k)

fig = plt.figure(figsize=(6,5))
plt.plot(k_grid, fig1, label=r'$\theta_\mathrm{eject}$')
plt.plot(k_grid, fig2, label=r'$\delta$')
plt.plot(k_grid, fig3, label=r'$\gamma$')
plt.plot(k_grid, fig4, label=r'$\log_{10}M_\mathrm{gaslimit}$')
plt.plot(k_grid, fig5, label=r'$\beta$')
plt.xscale('log')
plt.legend()
plt.xlabel(r'$k \; h/\mathrm{Mpc}$')
plt.show()
```

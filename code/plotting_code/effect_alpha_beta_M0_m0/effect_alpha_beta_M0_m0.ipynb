```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.size": 16,          # base font size
    "axes.labelsize": 16,     # x and y labels
    "axes.titlesize": 18,     # title
    "legend.fontsize": 14,    # legend
    "xtick.labelsize": 14,    # x tick labels
    "ytick.labelsize": 14,    # y tick labels
})

import sys
import os
import copy

sys.path.append(os.path.abspath("/home/milan/Desktop/thesis/code")) 
```

```python
from halo_model.power_spectra.param_effect_DMO import param_effect_spectrum_DMO
from halo_model.config.config import Config

sys.path.append(os.path.abspath("/home/milan/Desktop/thesis/code/plotting_code/effect_alpha_beta_M0_m0")) 

cfg = Config()
cfg.m_min = 1
cfg.M_min = 1
cfg.alpha = 0
```

```python
fig3 = param_effect_spectrum_DMO(cfg, 'alpha', -1)
```

```python
fig4 = param_effect_spectrum_DMO(cfg, 'beta', -1)
```

```python
fig5 = param_effect_spectrum_DMO(cfg, ['alpha', 'beta'], [-1, -1])
```

```python

k_grid = np.geomspace(cfg.k_min, cfg.k_max, cfg.N_k)

fig, ax = plt.subplots(figsize=(6,5))

ax.plot(k_grid, fig3, label=r'HMF')
ax.plot(k_grid, fig4, label=r'SHMF')
ax.plot(k_grid, fig5, label=r'HMF + SHMF')

ax.set_xlim(1e-1, 1e4)
ax.set_ylim(0.9, 1.005)
ax.set_xscale('log')
ax.legend(loc= 'upper right')

ax.set_ylabel(r'$P_{m, \rm mod}(k) / P_m(k)$')
ax.set_xlabel(r'$k \; h/\mathrm{Mpc}$')

textstr = (
    r'$M_0 = m_0 = 10^9 M_\odot$' 
)

ax.text(
    0.1, 0.2, textstr,
    transform=ax.transAxes,   
    fontsize=12,
    verticalalignment='top',
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
)
plt.tight_layout()
plt.savefig('alpha_beta_effect_comparison.pdf')
plt.show()
```
